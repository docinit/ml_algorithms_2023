1. Подберите скорость обучения (eta) и количество итераций

```python
import numpy as np
# 1. добавлены функции из урока
# 2. добавлены данные из урока

def calc_mae(y, y_pred):
    err = np.mean(np.abs(y - y_pred))
    return err

def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err

X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])

y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
# ИЗМЕНЕНИЯ:
# 3. изменен (удален) парамер eta перед началом цикла
n = X.shape[0]

# eta = 1e-2 - задается в функции
n_iter = int(1e3)

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(2,n_iter+2):
    eta = int(np.log2(i+1))**-2
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))

    if i % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)} eta = {round(eta,6)}')
```

3*. Вместо того, чтобы задавать количество итераций, задайте другое условие останова алгоритма - когда веса перестают изменяться меньше определенного порога $\epsilon$.

```python
# Добавлен критерий останова
n = X.shape[0]

n_iter = int(1e3)

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')
err_diff = np.inf # используется для остановки вычислений

for i in range(2,n_iter+2):
    eta = int(np.log2(i+1))**-2
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if abs(err_diff - err) < 1e-4: # работа функции будет остановлена, если разница между ошибками на разных шагах будет меньше 1e-4
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 4)} eta = {round(eta,6)}')
        break
    else:
        err_diff = err
    if i % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 4)} eta = {round(eta,6)}')
```
